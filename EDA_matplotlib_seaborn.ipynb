```python
#import dataset
import pandas as pd
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
#Line Plot -Total sales by region
df.groupby('Region')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
sales_data_agg = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(6,4))
plt.plot(sales_data_agg.index,sales_data_agg,marker='o')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
#plt.grid(True)
plt.show()
```

```python
#plot Total sales by region using groupby
df.groupby('Region')['Sales'].sum().plot(kind='line')
```

```python
#Bar plot for total sales by product
import matplotlib.pyplot as plt
product_data_agg = df.groupby('Product')['Sales'].sum()
plt.figure(figsize=(6,4))
plt.bar(product_data_agg.index,product_data_agg)
plt.title('Total Sales By Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
#plt.grid()
plt.xticks(rotation=45)
plt.show()
```

```python
#plot a histogram for distribution of price
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(6,4))
plt.hist(df['Price'],bins=np.linspace(50,1100,10),edgecolor='black')
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xticks(np.linspace(50,1100,10))
plt.grid()
plt.show()
```

```python
#Scatter plot of Sales vs Cost
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.scatter(df['Sales'],df['Cost'],color='red',alpha=0.5,marker='x')
plt.title('Sales vs Cost')
plt.xlabel('Sales')
plt.ylabel('Cost')
plt.grid()
plt.show()
```

```python
df.groupby('Region')['Sales'].sum()
```

```python
#Pie chart for sales by region
sales_by_region = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8, 6))
plt.pie(sales_by_region, labels=sales_by_region.index,
autopct='%1.1f%%', startangle=90,
colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'],
wedgeprops={'edgecolor':'black'})
plt.title('Sales Distribution by Region')
plt.show()
```

```python
#Box plot for sales by product
plt.figure(figsize=(10, 8))
df.boxplot(column='Sales', by='Product', patch_artist=True)
plt.title('Sales Distribution by Product')
plt.suptitle('')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
#sns.boxplot(x='Product',y='Sales',data=df)
sns.boxplot(x='Cost',data=df)
plt.title('Box plot')
plt.show()
```

```python
# Subplots: Line Plot, Bar Plot, and Scatter Plot
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
# Line plot
axs[0].plot(df['Price'], df['Sales'], marker='o', linestyle='-', color='blue')
axs[0].set_title('Sales vs Price')
axs[0].set_xlabel('Price')
axs[0].set_ylabel('Sales')
axs[0].grid(True)
# Bar plot
axs[1].bar(df['Product'], df['Sales'], color='skyblue')
axs[1].set_title('Sales by product')
axs[1].set_xlabel('Product')
axs[1].set_ylabel('Sales')
axs[1].tick_params(axis='x', rotation=45)
# Scatter plot
axs[2].scatter(df['Cost'], df['Sales'], color='red', alpha=0.7)
axs[2].set_title('Sales vs Cost')
axs[2].set_xlabel('Cost')
axs[2].set_ylabel('Sales')
axs[2].grid(True)
plt.tight_layout()
plt.show()
```
