BAR PLOT

gives you the frequency on the y axis

x axis should ideally be categorical in nature

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

Movie_type = ['Comedy','Action','Romance','Drama','Scifi']#x axis

Responses = [4,5,6,1,4]#y axis

ax.bar(Movie_type, Responses)

ax.set_ylabel("Survey Responses")

ax.set_xlabel("Movie_Type")

ax.set_title("MOVIES")

ax.set_xticklabels(Movie_type, rotation=30)

plt.show()
```

```python
import matplotlib.pyplot as plt
import statistics

fig, ax = plt.subplots()

Movie_type = ['Comedy','Action','Romance','Drama','Scifi']

Responses = [4,5,6,1,4]

Responses_std = statistics.stdev(Responses)

ax.bar(Movie_type, Responses,yerr=Responses_std)

ax.set_ylabel("Survey Responses")

plt.show()
```

```python
#plotting multiple data
import matplotlib.pyplot as plt

Movie_type = ['Comedy','Action','Romance','Drama','Scifi']

Men_responses = [4,5,6,1,4]

Women_responses = [8, 2, 7, 2, 1]

fig, ax = plt.subplots()

ax.bar(Movie_type, Men_responses, color = 'b', width = 0.50, label="Men Response")

ax.bar(Movie_type, Women_responses, color = 'g', width = 0.25, label="Women Response")

ax.set_ylabel("Survey Responses")

ax.legend()

ax.set_xticklabels(Movie_type, rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
Movie_type = ['Comedy', 'Action', 'Romance', 'Drama', 'Scifi']
Men_responses = [4, 5, 6, 1, 4]
Women_responses = [8, 2, 7, 2, 1]

# Positions for side-by-side bars
x = np.arange(len(Movie_type))
width = 0.35

# Plot
plt.figure()
plt.bar(x - width/2, Men_responses, width, label='Men Response')
plt.bar(x + width/2, Women_responses, width, label='Women Response')

# Labels and formatting
plt.ylabel('Survey Responses')
plt.xticks(x, Movie_type, rotation=45)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#Stacked bar chart
import matplotlib.pyplot as plt
Movie_type = ['Comedy','Action','Romance','Drama','Scifi']

Men_responses = [4,5,6,1,4]

Women_responses = [8, 2, 7, 2, 1]


fig, ax = plt.subplots()

ax.bar(Movie_type, Men_responses, color = 'b')

ax.bar(Movie_type, Women_responses, bottom = Men_responses, color = 'r')


ax.set_ylabel("Survey Responses")
ax.set_xticklabels(Movie_type, rotation=45)

ax.legend(labels=['Men', 'Women'])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)
#It takes three arguments, mean and standard deviation of the normal distribution, and the number of values desired.

## combine these different collections into a list
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

fig, ax = plt.subplots()

# Create the boxplot
ax.boxplot(data_to_plot)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)
#histogram will convert your data into class intervals: x axis
#y axis will show the frequency
#univariate plot

plt.hist(x, bins=10, histtype="stepfilled")#bins represent class interval
plt.show()
```
