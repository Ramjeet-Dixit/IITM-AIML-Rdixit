```python
x = iris.data #input array assigned as x
y = iris.target #output variable assigned as y
```

```python
#split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=123)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=8)
#assume k = 8

knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
#predict the labels on the test data
```

```python
y_pred
```

```python
y_test
#pre existing or the actual labels
```

```python
x_test.shape
```

```python
#number (%) of correct predictions; number of labels correctly predicted
knn.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix


print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
y_pred_train = knn.predict(x_train)
#predict the labels on the test data
```

```python
print(classification_report(y_train, y_pred_train))
```

### Tune and Optimize kNN in Python Using scikit-learn

```python
from sklearn.model_selection import GridSearchCV

parameters = {"n_neighbors": range(1, 50)}

gridsearch = GridSearchCV(KNeighborsClassifier(), parameters, cv=5)

gridsearch.fit(x_train, y_train)
```

GridSearchCV repeatedly fits kNN classifers on a part of the data and tests the performances on the remaining part of the data.

Doing this repeatedly will yield a reliable estimate of the predictive performance of each of the values for k.

In this example, you test the values from 1 to 50.

In the end, it will retain the best performing value of k, which you can access with .best_params_:

```python
gridsearch.best_params_
```

In this code, you print the parameters that have the lowest error score.

With .best_params_, you can see that choosing 3 as value for k will yield the best predictive performance.

Now that you know what the best value of k is, you can see how it affects your train and test performances:

```python
y_pred = gridsearch.predict(x_test)
#n_neighbors=3
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix


print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
y_pred_train = gridsearch.predict(x_train)
print(classification_report(y_train, y_pred_train))
```

**Improvising results using weights hyperparamter**

```python
parameters={'n_neighbors': range(1, 50),
                         'weights': ['uniform', 'distance']}
GD = GridSearchCV(KNeighborsClassifier(),parameters,cv=5)
# default options --> uniform (all points in the neighborhood are weighted equally)
# another option --> distance (weights closer neighbors more heavily than farther neighbors)
```

```python
GD.fit(x_train, y_train)
```

```python
GD.best_params_
```

```python
y_pred = GD.predict(x_test)
```

```python
print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
y_predtrain = GD.predict(x_train)
```

```python
print(confusion_matrix(y_train, y_predtrain))

print(classification_report(y_train, y_predtrain))
```
