## 7. Collections: List, Tuple, Set, Dictionary

```python
# List
subjects = ["Math", "Python", "DS"]
subjects.append("ML")
print(subjects)

# Tuple
grades = (90, 85, 92)
print(grades)

# Set
unique_ids = {101, 102, 103}
print(unique_ids)

# Dictionary
student = {"name": "IITM", "age": 21}
print(student["name"])
```

## 8. Conditional Statements

```python
marks = 70
if marks >= 75:
    print("Distinction")
elif marks >= 50:
    print("Pass")
else:
    print("Fail")
```

## 9. Loops (for, while)

```python
# for loop
for i in range(5):
    print(i)

# while loop
i = 0
while i < 5:
    print(i)
    i += 1
```

## 10. Functions

## 11. Classes and Objects

```python
class Student:
    def __init__(self, name, id):
        self.name = name
        self.id = id

    def show(self):
        print(f"Name: {self.name}, ID: {self.id}")

s1 = Student("Ravi", 1001)
s1.show()
```

## âœ… Practice Challenge
Write a function that calculates the factorial of a number.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```
